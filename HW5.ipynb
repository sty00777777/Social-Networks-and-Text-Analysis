（1）构建词袋模型，将代码贴于下方：

```python
# 初始化部分

# 选取的是刘慈欣的小说三体之第一部之地球往事
import pandas as pd
# 读入为数据框，一行为一段。\u3000是全角分割符号
raw = pd.read_csv("data\三体第一章_地球往事.txt",names=['txt'],sep='\u3000\u3000',encoding='gbk')
display(raw.head())

# 初始化分词库
# 设定分词及清理停用词函数
stoplist = list(pd.read_csv('data/停用词表.txt', names = ['w'], sep = 'aaa',encoding = 'utf-8', engine='python').w)
# 构建分词工具函数
import jieba
jieba.load_userdict('data\三体词库.txt')
def m_cut(intxt):
    return [ w for w in jieba.cut(intxt) if w not in stoplist and len(w) > 1 ]
```

```python
# 进行章节标注
def m_head(tmpstr):
    return tmpstr[:1] # 提取首个字符
def m_mid(tmpstr):
    return tmpstr.find("章 ") # 返回包含'回 '的位置

raw['head'] = raw.txt.apply(m_head)
raw['mid'] = raw.txt.apply(m_mid)
chapnum = 0

for i in range(len(raw)):
    if raw['head'][i] == '第' and raw['mid'][i] > 0:
        chapnum +=1
    raw.loc[i,'chap'] = chapnum
del raw['head']
del raw['mid']
raw.head()
```

```python
# 还原各章节
chapter = raw.groupby('chap').agg(sum) # 列为字符串时，sum函数自动转为合并字符串
# len(gp) # 小说共40回
display(chapter.txt[1]) # 第一回
display(chapter.head())
```

```python
# 分词后，将每个单词连接，中间用空格
rawchap = [ " ".join(m_cut(w)) for w in chapter.txt]
rawchap[0]
```

```python
len(rawchap)
```

```python
#主题词提取
import jieba.analyse
data = pd.DataFrame(columns=['章节','关键词'])
for i in range(len(rawchap)):
    text = rawchap[i]
    keywords = jieba.analyse.extract_tags(text, topK = 5, withWeight = False, allowPOS = ())
    print(i+1,'章节关键词为',keywords)
```

```python
# 构建词袋模型
from sklearn.feature_extraction.text import CountVectorizer
countvec = CountVectorizer(min_df = 5) # 在5个以上章节中出现的才保留
res = countvec.fit_transform(rawchap)

print(res.shape)
# 行为章节数，列为词汇数量
a = countvec.get_feature_names()
print('单词名称为：',a)

#基于词频矩阵X计算TF-IDF值 
from sklearn.feature_extraction.text import TfidfTransformer
transformer = TfidfTransformer() 
tfidf = transformer.fit_transform(res) 
tfidf
```

```python
# 文本聚类
# 利用k-means进行聚类分析
from sklearn.cluster import KMeans 
clf = KMeans(n_clusters = 5) 
s = clf.fit(tfidf) 
print(s)
print(clf.cluster_centers_.shape)
clf.labels_
```

```python
chapter['clsres'] = clf.labels_
chapter.tail()
```

```python
# 自动摘要
# 拆句子
def cut_sentence(intxt): 
    delimiters = frozenset('。！？') 
    buf = [] 
    for ch in intxt: 
        buf.append(ch) 
        if delimiters.__contains__(ch): 
            yield ''.join(buf) 
            buf = [] 
    if buf: 
        yield ''.join(buf) 
 
sentdf = pd.DataFrame(cut_sentence(chapter.txt[1]))
sentdf
```

```python
# 去除过短的句子，避免摘要出现无意义的内容
sentdf['txtlen'] = sentdf[0].apply(len)
sentlist = sentdf[0][sentdf.txtlen > 20]
print(len(sentlist))
sentlist.head()
```

```python
import jieba
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
txtlist = [ " ".join(jieba.lcut(w)) for w in sentlist]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(txtlist) # 将文本中的词语转换为词频矩阵
```

```python
tfidf_matrix = TfidfTransformer().fit_transform(X)
```

```python
# 利用nx包实现pagerank算法
import networkx as nx
similarity = nx.from_scipy_sparse_matrix(tfidf_matrix * tfidf_matrix.T) 
scores = nx.pagerank(similarity)
dict(list(scores.items())[0:10])
```

#### 计算两个句子之间的相似度

```python
import numpy as np
# 计算两个句子之间的相似度

# 定义目标文本
where = 10
target_text = rawchap[where]

# 转换为tf-idf向量
target_vec = transformer.transform(countvec.transform([target_text])).toarray()
# 初始化最大相似度和最相似的文本索引
max_sim = -1
max_index = -1

# 遍历所有的文本
for i in range(len(rawchap)):
    if i != where:
        # 获取当前文本对应的tf-idf向量
        vec = tfidf[i].toarray()
        # 计算余弦距离
        cosine_dist = np.dot(vec, target_vec.T) / (np.sum(vec**2)**0.5 * np.sum(target_vec**2)**0.5)
        # 如果余弦距离大于最大相似度，更新最大相似度和最相似的文本索引
        if cosine_dist > max_sim:
            max_sim = cosine_dist
            max_index = i

# 输出最相似的文本和相似度
print('与第',where+1,'章最相似的章节是第', max_index+1,'章')
print('相似度为:', max_sim)
```

#### 计算两个词语之间的相似度

```python
# 假设已经有了 D2M 矩阵 res.todense()
d2m_matrix = res.todense()

# 输入要查询相似度的词语
input_word = "丁仪"

# 获取输入词语在词袋模型中的索引位置
word_index = countvec.vocabulary_.get(input_word)

if word_index is None:
    print("词语不在词袋模型中")
else:
    # 提取输入词语的列向量
    input_vector = d2m_matrix[:, word_index]
    similar_cal_list = []

    for i in range(d2m_matrix.shape[1]):
        # 计算输入词语与其他词语的相似度
        # 计算向量的点积
        dot_product = np.dot(input_vector.T,d2m_matrix[:,i])

        # 计算向量的范数
        norm_A = np.linalg.norm(input_vector)
        norm_B = np.linalg.norm(d2m_matrix[:,i])

        # 计算余弦相似度
        similarities_cal = dot_product / (norm_A * norm_B)

        similar_cal_list.append(float(similarities_cal))

    # 使用argsort()函数获取按照元素大小升序排列的索引值
    sorted_indices = np.argsort(similar_cal_list)
    # 获取最大的六个值的索引并去掉最大的1对应的自己
    top_six_indices = sorted_indices[-6:-1]

    print("输入词语：", input_word)
    # 循环输出最大的五个词语
    for j in range(5):
        print('相似度排名为第'+str(j+1)+'的词语为'+list(countvec.vocabulary_.keys())[list(countvec.vocabulary_.values()).index(top_six_indices[4-j])]+'，相似度为' + str(similar_cal_list[top_six_indices[4-j]]))
```

（2）构建2-gram模型，将代码贴于下方：

二元模型（bigram model）

```python
# 数据预处理，将分词好的文章连接在一起，并且将其split为许多元素
whole_story = ''
for i in range(len(rawchap)):
    whole_story = whole_story + rawchap[i]

whole_story_split = whole_story.split(' ')
```

```python
from nltk.util import ngrams
from collections import Counter

bigram = ngrams(whole_story_split, 2)
bigram_count = Counter(bigram)

bigram_count.most_common()
```

```python
# 搜索特定词语并按频数输出二元组合
target_word = '丁仪'

# 根据频数进行排序
sorted_bigrams = sorted(
    bigram_count.items(),
    key=lambda x: x[1],
    reverse=True
)

# 取最高的五个二元组
top_five = []
for bigram, count in sorted_bigrams:
    if target_word in bigram:
        top_five.append((bigram, count))
        if len(top_five) >= 5:
            break

# 输出结果
for bigram, count in top_five:
    print(f"({bigram[0]}, {bigram[1]}): {count}")
```

```python
# 导入nltk.lm模块
from nltk.lm import MLE
from nltk.lm.preprocessing import padded_everygram_pipeline

# 使用padded_everygram_pipeline函数对数据进行预处理，生成填充符和N-gram序列，这里使用2-gram模型
n = 2
train_data, padded_sents = padded_everygram_pipeline(n, whole_story)

# 使用nltk.lm.Vocabulary类创建一个词汇表，用于限制模型的词汇范围
from nltk.lm import Vocabulary
vocab = Vocabulary([word for sent in padded_sents for word in sent])

# 使用nltk.lm.MLE类创建一个2-gram模型，并用训练数据进行训练
model = MLE(n)
model.fit(train_data, vocab)
```

```python
# 使用nltk.lm.api模块提供的方法对单个句子进行评估，计算句子合理性
sentence = ['三体','文明', '毁灭', '人类']
print(model.score(sentence[-1], sentence[:-1])) # 计算最后一个词的概率
print(model.entropy(sentence)) # 计算句子的熵
print(model.perplexity(sentence)) # 计算句子的困惑度
```

```python
from gensim.models.word2vec import Word2Vec
n_dim = 300 # 指定向量维度，大样本量时300~500较好
w2vmodel = Word2Vec(vector_size = n_dim, min_count = 10) # size改为vector_size
w2vmodel.build_vocab(chapter.cut) # 生成词表
w2vmodel
```
